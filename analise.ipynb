O fator de qualidade mede quão “pura” é a ressonância, ou seja, quão pouco o sistema perde energia por ciclo.


Calcular o fator de qualidade Q:

Q = f0 / (fsup - finf)


f0 = frequência de pico

fsup - finf = largura de banda à meia altura (-3dB)

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

plt.style.use('seaborn-v0_8-darkgrid')  # Estilo mais moderno

def ler_dados(file_path):
    with open(file_path, "r") as file:
        lines = file.readlines()
    data = []
    pattern = re.compile(r"^\s*([\d\.]+)\s+([-]?\d+\.\d+)")
    for line in lines:
        match = pattern.match(line)
        if match:
            freq = float(match.group(1))
            db = float(match.group(2))
            data.append((freq, db))
    return pd.DataFrame(data, columns=["Frequency (GHz)", "S21 (dB)"])

def plotar_resonador(df, titulo):
    max_point = df.loc[df["S21 (dB)"].idxmax()]
    half_max = max_point["S21 (dB)"] - 3
    df_below = df[df["S21 (dB)"] >= half_max]
    f_low = df_below["Frequency (GHz)"].min()
    f_high = df_below["Frequency (GHz)"].max()
    bandwidth = f_high - f_low

    plt.figure(figsize=(10, 6))
    plt.plot(df["Frequency (GHz)"], df["S21 (dB)"], label='S21 (dB)', color='navy')
    plt.axvline(x=max_point["Frequency (GHz)"], color='red', linestyle='--',
                label=f'Pico: {max_point["Frequency (GHz)"]:.7f} GHz')
    plt.axhline(y=half_max, color='green', linestyle='--', label='-3 dB')
    plt.axvline(x=f_low, color='purple', linestyle='--', label=f'F. inferior: {f_low:.7f} GHz')
    plt.axvline(x=f_high, color='purple', linestyle='--', label=f'F. superior: {f_high:.7f} GHz')
    plt.scatter([max_point["Frequency (GHz)"]], [max_point["S21 (dB)"]], color='red', zorder=5)
    plt.xlabel("Frequência (GHz)")
    plt.ylabel("S21 (dB)")
    plt.title(titulo)
    plt.legend(title=f'Largura de banda: {bandwidth:.3f} GHz')
    plt.tight_layout()
    plt.show()

```

**Ressonadores de 3GHZ**

```python
# Uso para o arquivo 3GHz cap 1fF dB:
file_path_3_1fF = r"/Users/valentinaspohr/Desktop/tudo/unicamp/2s:2024/rouxinol/dados/Results/res 3GHz cap 1fF dB.txt"
df_3_1fF = ler_dados(file_path_3_1fF)
plotar_resonador(df_3_1fF, "Ressonador de 3 GHz com capacitor de 1 fF")

# Uso para o arquivo 3GHz cap 5fF dB:
file_path_3_5fF = r"/Users/valentinaspohr/Desktop/tudo/unicamp/2s:2024/rouxinol/dados/Results/res 3GHz cap 5fF dB.txt"
df_3_5fF = ler_dados(file_path_3_5fF)
plotar_resonador(df_3_5fF, "Ressonador de 3 GHz com capacitor de 5 fF")

# Uso para o arquivo 3GHz cap 51fF dB:
file_path_3_51fF = r"/Users/valentinaspohr/Desktop/tudo/unicamp/2s:2024/rouxinol/dados/Results/res 3GHz cap 51fF dB.txt"
df_3_51fF = ler_dados(file_path_3_51fF)
plotar_resonador(df_3_51fF, "Ressonador de 3 GHz com capacitor de 51 fF")
```
