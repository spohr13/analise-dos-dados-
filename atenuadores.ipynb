```python
import matplotlib.pyplot as plt
import numpy as np

# Load data from the text file
file_path = 'Non-Magnetic Cryogenic Attenuator 0dB.txt'  # Adjust path if needed
with open(file_path, 'r') as file:
    data = file.read().strip().split()

# Split into frequencies and attenuations (assuming first half is frequencies, second half is attenuations)
n = len(data) // 2
frequencies = [float(x) for x in data[:n]]
attenuations = [float(x) for x in data[n:]]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(frequencies, attenuations, label='Attenuation (dB)', color='blue', linewidth=1.5)
plt.xlabel('Frequency (GHz)')
plt.ylabel('Attenuation (dB)')
plt.title('Non-Magnetic Cryogenic Attenuator (0dB) Response')
plt.grid(True)
plt.legend()
plt.show()
```
